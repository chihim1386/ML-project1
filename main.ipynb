```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
data
```

```python
data.info()
```

```python
data.dropna().info() #drop null value
```

```python
data.dropna(inplace = True) #drop null value and replace the old data
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split 

X = data.drop(['median_house_value'], axis=1)
Y = data['median_house_value']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
train_data = X_train.join(Y_train)
```

```python
train_data 
```

```python
train_data.hist(figsize=(15,8))
```

```python
# Assuming train_data is already defined and is a DataFrame
# Calculate the correlation matrix while ignoring non-numeric columns
correlation_matrix = train_data.corr(numeric_only=True)

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")

plt.title("Correlation Heatmap")
plt.show()
```

```python
train_data["total_rooms"] = np.log(train_data["total_rooms"] + 1)
train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"] + 1)
train_data["population"] = np.log(train_data["population"] + 1)
train_data["households"] = np.log(train_data["households"] + 1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
pd.get_dummies(train_data.ocean_proximity)
```

```python
pd.get_dummies(train_data.ocean_proximity).astype(int) #change T/F to 1 and 0
```

```python
#join and drop ocean_proximity
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity).astype(int)).drop(["ocean_proximity"],axis=1)
```

```python
train_data
```

```python

plt.figure(figsize=(15, 8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.scatterplot(x='latitude', y='longitude', data=train_data, hue='median_house_value', palette="coolwarm")
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data["household_rooms"] = train_data["total_rooms"] / train_data["households"]
```

```python

plt.figure(figsize=(15, 8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train, Y_train = train_data.drop(['median_house_value'], axis=1), train_data['median_house_value']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression().fit(X_train_s, Y_train)
```

```python
test_data = X_test.join(Y_test)

test_data["total_rooms"] = np.log(test_data["total_rooms"] + 1)
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"] + 1)
test_data["population"] = np.log(test_data["population"] + 1)
test_data["households"] = np.log(test_data["households"] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity).astype(int)).drop(["ocean_proximity"],axis=1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data["household_rooms"] = test_data["total_rooms"] / test_data["households"]
```

```python
X_test, Y_test = test_data.drop(['median_house_value'], axis=1), test_data['median_house_value']
```

```python
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s, Y_test)
```
